```python
import requests

# Função de requisição à API da NASA para obter os dados dos asteroides
def pesquisar_asteroides(data):
  api_key = "DEMO_KEY"
  url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date={data}&end_date={data}&api_key={api_key}"

  resposta = requests.get(url)

  if resposta.status_code == 200:
    dados = resposta.json()
    return dados
  else:
    return {'Erro': 'Não foi possivel obter dados'}

# Função para processar os dados e armazená-los em um dicionário.
def armazenar_dic(dados,data):
  asteroides={}
  for i in range(len(dados['near_earth_objects'][data])):
    nome=(dados['near_earth_objects'][data][i]['name'])
    diametro_max=(dados['near_earth_objects'][data][i]['estimated_diameter']['meters']['estimated_diameter_max'])
    velocidade_rel = (dados['near_earth_objects'][data][i]['close_approach_data'][0]['relative_velocity']['kilometers_per_hour'])
    distancia = (dados['near_earth_objects'][data][i]['close_approach_data'][0]['miss_distance']['kilometers'])
    perigoso = (dados['near_earth_objects'][data][i]['is_potentially_hazardous_asteroid'])
    asteroides[nome] = {'diametro_max':diametro_max, 'velocidade_rel':velocidade_rel, 'distancia':distancia, 'perigoso':perigoso}
  return asteroides

# Função para exibir as informações formatadas para o usuário.
def exibir_pesquisa(asteroides,data):
  print(f'\n---Asteroides proximos a Terra em {data}---\n')
  print('-' * 55)
  for asteroide,valores in asteroides.items():
    print(f"Nome do asteroide                 : {asteroide}")
    print(f"Diametro Máximo [m]               : {asteroides[asteroide]['diametro_max']}")
    print(f"Velocidade Relativa a Terra [km/h]: {asteroides[asteroide]['velocidade_rel']}")
    print(f"Distancia Relativa a Terra [km]   : {asteroides[asteroide]['distancia']}")
    if asteroides[asteroide]['perigoso'] == True:
      print(f"Potencialmente Perigoso           : SIM! Perigo! Perigo!")
    else:
      print(f"Potencialmente Perigoso           : Não")
    print('-' * 55)

# Principal
continuar = 's'
while continuar == 's':
  data = input('Digite a data no formato aaaa-mm-dd: ')
  dados = pesquisar_asteroides(data)
  asteroides = armazenar_dic(dados,data)
  exibir_pesquisa(asteroides, data)
  continuar = input('Deseja continuar? [s/n]')





```
